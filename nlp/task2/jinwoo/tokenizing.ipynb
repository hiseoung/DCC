```python
import pandas as pd
from konlpy.tag import Okt
import os
```

```python
data_path = '../../../data/뉴스'

train_path = os.path.join(data_path, 'train_data.csv')
test_path = os.path.join(data_path, 'test_data.csv')
topic_path = os.path.join(data_path, 'topic_dict.csv')
sub_path = os.path.join(data_path, 'sample_submission.csv')

train = pd.read_csv(train_path, index_col=0)
test = pd.read_csv(test_path, index_col=0)
topic = pd.read_csv(topic_path)
sub = pd.read_csv(sub_path, index_col=0)
```

```python
train.info()
```

```python
def check_tag(pos_list):
    return_list = []
    for pos in pos_list:
        if pos[1] not in ['Josa', 'Punctuation', 'Unknown', 'KoreanParticle', 'Hashtag', 'ScreenName',
                         'Email', 'URL']:
            return_list.append(pos[0])
    
    return ' '.join(return_list)
```

```python
okt = Okt()
filtered_title = []
for title in train['title'].tolist():
    temp = check_tag(okt.pos(title))
    filtered_title.append(temp)
```

```python
token_df = pd.DataFrame({
    'title': filtered_title,
    'topic_idx': train['topic_idx'].to_list
})
token_df
```

```python
save_path = os.path.join(data_path, 'token.csv')
token_df.to_csv(save_path, index=False, encoding='utf-8-sig')
```

```python
filtered_title_test = []
for title in test['title'].tolist():
    temp = check_tag(okt.pos(title))
    filtered_title_test.append(temp)
```

```python
filtered_title_test
```

```python
test_token_df = pd.DataFrame({
    'title': filtered_title_test
})
save_path = os.path.join(data_path, 'test_token.csv')
test_token_df.to_csv(save_path, index=False, encoding='utf-8-sig')
```
